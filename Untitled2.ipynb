```python
import numpy as np

# Define two sample arrays
array1 = np.array([10, 20, 30, 40])
array2 = np.array([15, 20, 25, 35])

# Perform element-wise comparisons
greater = np.greater(array1, array2)
greater_equal = np.greater_equal(array1, array2)
less = np.less(array1, array2)
less_equal = np.less_equal(array1, array2)

# Print the results
print("Array1:", array1)
print("Array2:", array2)
print("Greater:", greater)
print("Greater Equal:", greater_equal)
print("Less:", less)
print("Less Equal:", less_equal)
```

```python
import numpy as np

# Define two sample arrays
array1 = np.array([10, 20, 30, 40])
array2 = np.array([15, 20, 25, 35])

# Perform element-wise comparisons
greater = np.greater(array1, array2)
greater_equal = np.greater_equal(array1, array2)
less = np.less(array1, array2)
less_equal = np.less_equal(array1, array2)

# Print the results
print("Array1:", array1)
print("Array2:", array2)
print("Greater:", greater)
print("Greater Equal:", greater_equal)
print("Less:", less)
print("Less Equal:", less_equal)

# Create arrays of zeros, ones, and fives
zeros_array = np.zeros(10, dtype=int)
ones_array = np.ones(10, dtype=int)
fives_array = np.full(10, 5, dtype=int)

# Print the arrays
print("Zeros Array:", zeros_array)
print("Ones Array:", ones_array)
print("Fives Array:", fives_array)

# Create an array of integers from 30 to 70
int_array = np.arange(30, 71)

# Print the array
print("Array of integers from 30 to 70:", int_array)

# Multiply the values of two given vectors
vector1 = np.array([1, 2, 3, 4])
vector2 = np.array([5, 6, 7, 8])
product = np.multiply(vector1, vector2)

# Print the product
print("Vector1:", vector1)
print("Vector2:", vector2)
print("Product of the two vectors:", product)
```
